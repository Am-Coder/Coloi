```python
import tensorflow as tf
import cv2
import numpy as np
import PIL
from tensorflow.keras import Sequential
from tensorflow.keras.layers import *
from PIL import Image
from tensorflow.keras.models import model_from_json
```

```python

img = Image.open('flower.jpg')

img = img.resize((400,400),Image.ANTIALIAS)
# img_rgb = cv2.cvtColor(np.array(img,'uint8'),cv2.COLOR_GRAY2RGB)
img_lab =  cv2.cvtColor(np.array(img,'uint8'),cv2.COLOR_RGB2BGR)
img_lab =  cv2.cvtColor(np.array(img,'uint8'),cv2.COLOR_BGR2Lab)


    
    
X=img_lab[:,:,0]
Y=img_lab[:,:,1:]
print(Y.shape)
X = X.reshape(1, 400, 400, 1)
Y = Y.reshape(1, 400, 400, 2)
```

```python
model = Sequential()

#Building network
model.add(InputLayer(input_shape=(None,None,1)))
model.add(Conv2D(8, (3,3),activation='relu',padding='same',strides=2))
model.add(Conv2D(8, (3,3), activation='relu',padding='same'))
model.add(Conv2D(16, (3,3),activation='relu',padding='same'))
model.add(Conv2D(16, (3,3),activation='relu',padding='same',strides=2))
model.add(Conv2D(32, (3,3),activation='relu',padding='same'))
model.add(Conv2D(32, (3,3),activation='relu',padding='same',strides=2))
model.add(UpSampling2D((2,2)))
model.add(Conv2D(32, (3,3),activation='relu',padding='same'))
model.add(UpSampling2D((2,2)))
model.add(Conv2D(16, (3,3),activation='relu',padding='same'))
model.add(UpSampling2D((2,2)))
model.add(Conv2D(2, (3,3),activation='relu',padding='same'))

model.compile(optimizer='rmsprop',loss='mse')


model.fit(x=np.array(X), y=np.array(Y), batch_size=1, epochs=2000)
print(model.evaluate(X,Y,batch_size=1))
```

```python
#output colorizations

# serialize model to JSON
model_json = model.to_json()
with open("coloi1.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("coloi1.h5")
print("Saved model to disk")


out = model.predict(X)
canvas = np.zeros((400, 400,3),'uint8')
canvas[:,:,0] = X[0][:,:,0]
canvas[:,:,1:] = out[0]
res = cv2.cvtColor(canvas,cv2.COLOR_Lab2BGR)
cv2.imshow('image',res)
cv2.imshow('image1',np.array(img_lab,'uint8'))
k=cv2.waitKey(0)
if k==ord('q'):
    cv2.destroyAllWindows()
```
